```python
import pandas as pd
```

```python
df = pd.read_csv("train.csv")
```

```python
df
```

```python
df.columns
```

```python
X = df.iloc[:,:-1]
```

```python
y = df['SalePrice']
```

```python
X
```

```python
X.isnull().sum()[X.isnull().sum()>0]
```

```python
missing_percentage = (X.isnull().sum()[X.isnull().sum()>0]/X.shape[0])*100
missing_percentage
```

```python
X = X.drop(missing_percentage[missing_percentage>50].index, axis=1)
```

```python
missing_percentage = (X.isnull().sum()[X.isnull().sum()>0]/X.shape[0])*100
missing_percentage
```

```python
features_with_missing = missing_percentage.index.to_list()
features_with_missing
```

```python
X['GarageCond'].value_counts()
```

```python
for feature in features_with_missing:
  if X[feature].dtype=='object':
    X[feature] = X[feature].fillna(X[feature].value_counts().index[0])
  else:
    X[feature] = X[feature].fillna(X[feature].mean())
```

```python
X.isnull().sum()[X.isnull().sum()>0]
```

```python
categorical_features = X.select_dtypes('object').columns.to_list()
categorical_features
```

```python
one_hot_encoded = pd.get_dummies(X[categorical_features])
one_hot_encoded
```

```python
X = X.drop(categorical_features, axis=1)
X = X.join(one_hot_encoded)
```

```python

X = X.drop(['Id'], axis=1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred = lr.predict(X_test)
```
